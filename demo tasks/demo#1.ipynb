```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
data = pd.read_csv('../data/adult.data.csv')
```

```python
data.head()
```

```python
ages_bog = round(data[data['salary']=='>50K']['age'].mean())
std_bog = round(data[data['salary']=='>50K']['age'].std(),1)
ages_bed = round(data[data['salary']=='<=50K']['age'].mean())
std_bed = round(data[data['salary']=='<=50K']['age'].std(),1)
print('Средний возраст и среднеквадратичное отклонения богатых: {0} +- {1}\nСредний возраст и среднеквадратичное отклонения бедных: {2} +- {3}'.format(ages_bog,std_bog,ages_bed,std_bed))
```

```python
for race1, sub_df in data.groupby(['race','sex']):
    print(race1)
    print(sub_df['age'].max())
```

```python
for (race1,sex1),sub_df in data.groupby(['race','sex']):
    print(race1,sex1)
    print(sub_df['age'].max())
```

```python
data.groupby(['race','sex'])['age'].agg(np.max)
```

```python
max_load = data['hours-per-week'].max()
print(max_load)
print(data[data['hours-per-week']==max_load].shape[0])
round(data[data['hours-per-week']==max_load]['salary'].value_counts(normalize=True),2)
```

```python
japaneeze = data[data['native-country']=='Japan']
japaneeze.groupby('salary')['hours-per-week'].mean()
print(round(japaneeze[japaneeze['salary']=='>50K']['hours-per-week'].mean()))
print(round(japaneeze[japaneeze['salary']=='<=50K']['hours-per-week'].mean()))
```

```python
japaneeze.groupby(['salary',])['hours-per-week'].mean()
```
